```python
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')
```

```python
x = data
x.dropna(subset=['Selling_Price'],inplace=True)
```

```python
rem_labels = ['Product_id','Customer_name']
```

```python
x = data.drop(axis=1,labels = rem_labels)
```

```python
x.Loyalty_customer.replace(to_replace='Yes',value=1,inplace=True)
x.Loyalty_customer.replace(to_replace='No ',value=0,inplace=True)
```

```python
x.Product_Category.replace(to_replace='Pet_care',value=7,inplace=True)
x.Product_Category.replace(to_replace='Repair',value=8,inplace=True)
x.Product_Category.replace(to_replace='Child_care',value=0,inplace=True)
x.Product_Category.replace(to_replace='Cosmetics',value=1,inplace=True)
x.Product_Category.replace(to_replace='Hospitality',value=5,inplace=True)
x.Product_Category.replace(to_replace='Technology',value=9,inplace=True)
x.Product_Category.replace(to_replace='Organic',value=6,inplace=True)
x.Product_Category.replace(to_replace='Home_decor',value=4,inplace=True)
x.Product_Category.replace(to_replace='Educational',value=2,inplace=True)
x.Product_Category.replace(to_replace='Fashion',value=3,inplace=True)
```

```python
data_ins = x[['instock_date']]
```

```python
data_ins['instock_date'] = data_ins['instock_date'].str[:4]
```

```python
x['instock_date'] = data_ins['instock_date']
```

```python
x['instock_date'] = x['instock_date'].astype('int64')
```

```python
data_pos = x[x['Selling_Price']>0]
```

```python
np.abs(data_pos.corr())
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3, weights='uniform', metric='nan_euclidean')
data_filled=imputer.fit_transform(data_pos)
```

```python
X=data_filled[:,[1,2,6,9,10]]
```

```python
Y=data_filled[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.2, random_state=43)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
import xgboost as xg
```

```python
XGB = xg.XGBRegressor(objective ='reg:linear', n_estimators = 10, seed = 123, max_depth=9)
```

```python
XGB.fit(X,Y)
```

```python
pred = XGB.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error as MSE
```

```python
rmse = np.sqrt(MSE(y_test, pred)) 
print("RMSE : % f" %(rmse))
```
